```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
```

```python
#loading data into the system and checking slightly how our data looks like
data = pd.read_csv("/content/dataset.csv")
data
```

```python
#what are the number of rows and columns in our data
data.shape
```

```python
#EDA with the pandas profiling report
report = ProfileReport(data)
report
```

```python
data.info()
```

```python
#lets do some data cleaning

#deleting the loan id column since it wont in any way used for analysis
data.drop(['Loan_ID'], axis=1, inplace=True)

#replacing the missing values in the columns of Gender, Dependents, Self_Employed,  Credit_History with median and Loan_Amount_Term with mean
data["Gender"].fillna(data['Gender'].mode()[0], inplace=True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean(), inplace=True)
```

```python
data.info()
```

```python
data['Loan_Status'].unique()
```

```python
#converting the categorical variables to numerical variable
#Gender
Gender = {'Male':1, 'Female':2}
data['Gender']=data['Gender'].map(Gender).astype(int)

#married
Married = {'Yes':1, 'No':0}
data['Married']=data['Married'].map(Married).astype(int)

#dependents
Dependents = {'1':1,'0':0,'2':2,'3+':3}
data['Dependents']=data['Dependents'].map(Dependents).astype(int)

#Education
Education = {'Graduate':1, 'Not Graduate':0}
data['Education']=data['Education'].map(Education).astype(int)

#self employed
Self_Employed = {'No':0, 'Yes':1}
data['Self_Employed']=data['Self_Employed'].map(Self_Employed).astype(int)

#property area
Property_Area = {'Rural':1, 'Urban':2, 'Semiurban':3}
data['Property_Area']=data['Property_Area'].map(Property_Area).astype(int)

#loan status
Loan_Status = {'N':0, 'Y':1}
data['Loan_Status']=data['Loan_Status'].map(Loan_Status).astype(int)
```

```python
#taking a look at our data
data.head()
```

```python
#creating a machine learning dataset
x=data.drop('Loan_Status', axis=1)
y=data['Loan_Status']
```

```python
#checking what x looks like
x
```

```python
#checking on what y looks like
y
```

```python
#splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(f'Training shape x:',x_train.shape)
print(f'Testing shape x:',x_test.shape)
print('****_______*****_______*****')
print(f'Training shape y:',y_train.shape)
print(f'Testing shape y:',y_test.shape)
```

```python
#normlizing our data
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
x_train = mm.fit_transform(x_train)
x_test = mm.transform(x_test)
```

```python
#train the support vector classification model
from sklearn.svm import SVC
#creating the SVM
svm = SVC(kernel='rbf', random_state=0, gamma=10, C=1.0)
svm.fit(x_train, y_train)
print("Train accuracy:", svm.score(x_train, y_train))
print("Test accuracy:", svm.score(x_test, y_test))
#predicted y values
y_pred = svm.predict(x_test)
print(y_pred)
#confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(f'Confusion matrix:', cm)
print(f'Accuracy:', accuracy_score(y_test, y_pred)*100, '%')
print(classification_report(y_test, svm.predict(x_test)))
```

```python
#training the xgboost model
import xgboost as xgb
#creating the xgboost model
xgb_clf = xgb.XGBClassifier()
xgb_clf = xgb_clf.fit(x_train, y_train)
print("Train accuracy:", xgb_clf.score(x_train, y_train))
print("Testing accuracy:", xgb_clf.score(x_test,y_test))
#predicted y values
y_pred = xgb_clf.predict(x_test)
print(y_pred)
#confusion matrix
cm = confusion_matrix(y_test,y_pred)
print(f'Confusion matrix:', cm)
print(f'Accuracy:', accuracy_score(y_test,y_pred)*100, '%')
print(classification_report(y_test, xgb_clf.predict(x_test)))
```

```python
#train a decision tree model
from sklearn import tree
#creatiing tree object
decision_tree = tree.DecisionTreeClassifier(criterion='gini')
decision_tree.fit(x_train, y_train)
print("Train accuracy:", decision_tree.score(x_train, y_train))
print("Test accuracy:", decision_tree.score(x_test, y_test))
#predicted values for y
y_pred = decision_tree.predict(x_test)
print(y_pred)
#confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(f'Confusion matrix:', cm)
print(f'Accuracy:', accuracy_score(y_test, y_pred)*100, '%')
print(classification_report(y_test, decision_tree.predict(x_test)))
```

```python
#training the random forest
from sklearn.ensemble import RandomForestClassifier
#creating a random forest object
random_forest = RandomForestClassifier()
#training the model
random_forest.fit(x_train, y_train)
print("Train accuracy:", random_forest.score(x_train, y_train))
print("Test accuracy:", random_forest.score(x_test, y_test))
#predicted values for y
y_pred = random_forest.predict(x_test)
print(y_pred)
#confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(f'Confusion matrix:', cm)
print(f'Accuracy:', accuracy_score(y_test, y_pred)*100, '%')
print(classification_report(y_test, random_forest.predict(x_test)))
```

```python
#feature importance for random forest classifier
rf = random_forest
rf.feature_importances_
sorted_idx = rf.feature_importances_.argsort()
plt.barh(data.columns[sorted_idx],rf.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```
